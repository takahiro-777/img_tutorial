```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
cat_cascade = cv2.CascadeClassifier('./cascade/haarcascade_frontalcatface_extended.xml')

img_ori = cv2.imread('./test_input/cat.jpg')
img = cv2.imread('./test_input/cat.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cat_faces = cat_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in cat_faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),8)
```
