```python
"""
参考リスト
https://www.blog.umentu.work/python3-opencv3で背景差分を求める/
-> 例としてわかりやすかったので上記記事を参考にした。
"""
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
face_cascade = cv2.CascadeClassifier('./cascade/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('./cascade/haarcascade_eye.xml')

img_ori = cv2.imread('./test_input/face01.jpg')
img = cv2.imread('./test_input/face01.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
```

```python
# 処理後全体画像の貼り付け
plt.imshow(cv2.cvtColor(img_ori, cv2.COLOR_BGR2RGB))
# 処理後全体画像の表示
plt.show()
```

```python
# 処理後全体画像の貼り付け
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
# 処理後全体画像の表示
plt.show()
```

```python
# 処理後画像の貼り付け
plt.imshow(cv2.cvtColor(roi_color, cv2.COLOR_BGR2RGB))
# 処理後画像の表示
plt.show()
```
