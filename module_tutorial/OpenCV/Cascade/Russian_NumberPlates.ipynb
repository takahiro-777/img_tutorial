```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
np_cascade = cv2.CascadeClassifier('./cascade/haarcascade_licence_plate_rus_16stages.xml')
number_cascade = cv2.CascadeClassifier('./cascade/haarcascade_russian_plate_number.xml')

img_ori = cv2.imread('./test_input/russia_car.jpg')
img = cv2.imread('./test_input/russia_car.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
number_plates = np_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in number_plates:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),8)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    numbers = number_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in numbers:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
```

```python
# 
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
# 
plt.show()
```
